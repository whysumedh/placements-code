```python
#Gender Classification
import pandas as pd
from naampy import predict_fn_gender

input_file = "gender.csv"
output_file = "output_with_gender.csv"

df = pd.read_csv(input_file)

if 'First_Name' not in df.columns:
    raise KeyError("The 'First_Name' column is missing in the input CSV.")

predictions = predict_fn_gender(df['First_Name'].tolist())

pred_df = pd.DataFrame(predictions, columns=["name", "pred_gender", "pred_prob"])

df['Gender'] = pred_df['pred_gender']
df['Gender_Probability'] = pred_df['pred_prob']

df.to_csv(output_file, index=False)

print(f"File saved successfully with gender predictions at: {output_file}")
```

```python
import pandas as pd

def get_ctc(row):
    if row['Company'] == 'Cognizant':
        if row['Package'] == 'GenC Select':
            return '4.0LPA'
        elif row['Package'] == 'GenC Next Select':
            return '6.75LPA'
    elif row['Company'] == 'TCS':
        if row['Package'] == 'Ninja':
            return '3.3LPA'
    elif row['Company'] == 'LTIMindtree':
        return '4.0LPA'
    elif row['Company'] == 'Infosys':
        return '3.6LPA'
    return None

def process_placement_data(input_file, output_file):
    df = pd.read_excel(input_file)

    df['CTC'] = df.apply(get_ctc, axis=1)

    # Save the modified DataFrame to a new CSV file
    df.to_excel(output_file, index=False)
    print(f"Processed data saved to {output_file}")

input_file = 'WITCH-P-1.xlsx'
output_file = 'WITCH-P-2.xlsx'
process_placement_data(input_file, output_file)
```

```python
import pandas as pd
df=pd.read_csv('WITCH-P.csv')
def preprocess_and_filter_dataframe(df):
    df['CTC'] = df['CTC'].fillna('').astype(str)
    df['CTC'] = df['CTC'].str.replace('LPA', '', regex=False)
    df['CTC'] = pd.to_numeric(df['CTC'], errors='coerce')
    df['Company_Priority'] = df['Company'].apply(lambda x: 1 if x == 'LTIMindtree' else 2)
    df = df.sort_values(by=['Reg_No', 'CTC', 'Company_Priority'], ascending=[True, False, True])
    df = df.loc[df.groupby('Reg_No')['CTC'].idxmax()].reset_index(drop=True)
    df = df.drop(columns=['Company_Priority'])
    return df
```
